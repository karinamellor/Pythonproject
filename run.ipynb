```python
import tkinter as tk

def numerology_value(letter):
    # Define a dictionary to map each letter to its numerology value
    numerology_map = {
        'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6, 'g': 7, 'h': 8, 'i': 9,
        'j': 1, 'k': 2, 'l': 3, 'm': 4, 'n': 5, 'o': 6, 'p': 7, 'q': 8, 'r': 9,
        's': 1, 't': 2, 'u': 3, 'v': 4, 'w': 5, 'x': 6, 'y': 7, 'z': 8
    }
    
    # Convert the letter to lowercase to match the keys in the numerology map
    letter = letter.lower()
    
    # Look up the numerology value for the letter
    if letter in numerology_map:
        return numerology_map[letter]
    else:
        return 0

def calculate_numerology_value(name):
    # Calculate the numerology value for the entire name
    numerology_value = 0
    for letter in name:
        numerology_value += numerology_value(letter)
    
    return numerology_value

def on_submit():
    # Get the name from the input box
    name = name_entry.get()
    
    # Calculate the numerology value of the name
    numerology_value = calculate_numerology_value(name)
    
    # Update the result label
    result_label.config(text=f"The numerology value of your name {name} is {numerology_value}.")
    

# Create the main window
root = tk.Tk()
root.title("Numerology Calculator")

# Create the label and input box for the name
name_label = tk.Label(root, text="Enter your name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

# Create the submit button
submit_button = tk.Button(root, text="Calculate", command=on_submit)
submit_button.pack()

# Create the result label
result_label = tk.Label(root, text="")
result_label.pack()

# Start the main loop
root.mainloop()
```
